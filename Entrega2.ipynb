**Para unir las 3 bases de datos, por las columnas que comparten que son Fecha, Gravedad y Clase de Accidente, se procede a crear una serie de tiempo para cada una de las ciudades por la fecha para asi estar seguros que se analizaran todas las fechas de cada base (df1,df2,df3)**

```python
df4= pd.DataFrame(pd.date_range(start=df1['FECHA'].min(), end=df1['FECHA'].max())).rename(columns ={0:'FECHA'}) ##Serie de Tiempo de la ciudad de Palmira
df4
```

**Se procede a unir la base de datos original con la serie de tiempo correspondiente, para esto se utiliza la funcion merge**

```python
df7= pd.merge(df4, df1, on ='FECHA', how ='left')##Palmira
```

```python
df8= pd.merge(df5, df2, on ='FECHA', how ='left')##Envigado
df9= pd.merge(df6, df3, on ='FECHA', how ='left')##Medellin
```

**Luego se procede a agrupar la cantidad de accidentes por la fecha, gravedad y la clase de accidente y se hace un count para contar la cantidad de accidentes que hubo**

**Se agrupan las tres bases de datos con el merge para observar la cantidad de accidentes que hubieron en la ciudad de Palmira, Envigado y Medellin dependiendo de la fecha**

```python
bodega=pd.merge(palmira,envigado, on=['FECHA','GRAVEDAD','CLASE_ACCIDENTE'], how='outer').merge(medellin, on=['FECHA','GRAVEDAD','CLASE_ACCIDENTE'], how='outer').rename(columns ={'JORNADA':'PALMIRA','RADICADO':'ENVIGADO','NUMCOMUNA':'MEDELLIN'})
```

```python
df9=df9.rename(columns ={'DISEÃ\x91O':'DISEÑO'})  #Se renombra el titulo de la columna 
```

```python
df9['DISEÑO']=df9['DISEÑO'].replace({'PontÃ³n':'Ponton'})  #se reemplaza el nombre
```

```python
#se crea un dataframe con diseño y lo cuento con expediente, ordeno de manera descendente
tramo_via = df9.groupby(['DISEÑO'])[['EXPEDIENTE']].count().sort_values('EXPEDIENTE', ascending = False).rename(columns={'EXPEDIENTE':'ACCIDENTES'})
tramo_via['porcentaje'] = tramo_via.apply(lambda x: x.cumsum()/tramo_via['ACCIDENTES'].sum()) #Agregamos columna con el porcentaje acumulado segun la frecuencia de accidentes
tramo_via
```

# INFORMACION ADICIONAL Medellín

```python
##Base accidentalidad en el municipio de Medellin para el analisis adicional
dfm = pd.read_csv('http://medata.gov.co/sites/default/files/medata_harvest_files/mede_victimas_inci.csv',sep=';',nrows= 200000, usecols = [1,2,5,14])
#se tomaron solo 200.000 filas porque si se toman todas las filas las ultimas las carga como NaN
#se tomaron solo la columna 1,2 y 5  porque son las que tienen las categorías relevantes para complementar nuestro análisis
#se toma la columna de radicado para poder realizar las gráficas
dfm
```

**Se tomará la hora de los accidentes, pero solo se tendrá en cuenta para la visualización las primeras 10 horas que más accidentes presenta**

NOTA: SE UTILIZARÁN LOS DATOS QUE VAN DESDE EL 01/01/2014 HASTA EL 22/05/2020.






```python
hora_acc = dfm.groupby(['Hora_incidente'])[['Radicado']].count().reset_index().rename(columns={'Radicado':'Cantidad'})
hora_acc = hora_acc.head(10) #visualizar mejor
```

```python
# definir gráfica
fig = px.bar(hora_acc, x='Hora_incidente', y=['Cantidad'], barmode= 'group', title ='<b>Accidentes según la hora<b>')
px.bar()

# agregar detalles a la gráfica
fig.update_layout(
    xaxis_title = 'Hora',
    yaxis_title = 'Cantidad de Accidentes',
    template = 'simple_white',
    title_x = 0.5,
    legend_title = '<b>variable<b>')


fig.show()
```

**Teniendo en cuenta el rango de hora que se tomó (top 10) por la cantidad de datos arrojo que la accidentalidad en Medellín se presenta entre las horas 00 y las 9am, donde las 00 horas en donde más accidentalidad se presenta con 422, donde una posible causa de esto es que a esta hora mucha gente no respeta los semáforos o se confía de la vía tan despejada y andan a una alta velocidad, a su vez la hora 5 am es la que le sigue con una accidentalidad de 262, se puede dar porque a esta hora mucha gente se dirige a su lugar de trabajo y pueden incurrir en excesos de velocidad.**

**También se tomará el sexo para complementar el análisis.**

```python
dfm['Sexo']= dfm['Sexo'].str.upper() # se pone todo en mayúscula para que todo quede en el mismo formato
```

```python
sex_acc = dfm.groupby(['Sexo'])[['Radicado']].count().reset_index().rename(columns={'Radicado':'Cantidad'})

#Se realiza el grafico de dataframe total
fig = px.pie(sex_acc , values = 'Cantidad', names = 'Sexo', title = '<b>% Accidentes según el sexo<b>')
# agregar detalles a la gráfica
fig.update_layout(
    template = 'simple_white',
    legend_title = 'Accidentalidad por Sexo',
    title_x = 0.5)

fig.show()
```


**El sexo que más tuvo accidentes de tránsito fue el masculino con un 68.7% superando por un poco más del doble al sexo contrario, muy pocos datos no cuentan con esta categoría ya que solo es un 0.62% de estos.**
